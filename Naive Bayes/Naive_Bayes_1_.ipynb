```python
import pandas as pd
```

```python
df=pd.read_csv('/content/titanic-2.csv')
df.head()
```

```python
df1=df.drop(['PassengerId','Name','SibSp','Parch','Ticket','Cabin','Embarked'],axis='columns')
df1.head()
```

```python
inputs=df1.drop('Survived',axis='columns')
target=df1['Survived']
```

```python
dummies=pd.get_dummies(inputs.Sex)
dummies.head(3)
```

```python
final=pd.concat([inputs,dummies],axis='columns')
final.head(3)
```

I am dropping male column as well because of dummy variable trap theory. One column is enough to repressent male vs female

```python
final.drop(['Sex','male'],axis='columns',inplace=True)
final.head(3)
```

```python
final.columns[final.isna().any()]
```

```python
final[:10]
```

```python
final.Age=final.Age.fillna(final.Age.mean())
final.head()
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(final,target,test_size=0.2)
```

```python
from sklearn.naive_bayes import GaussianNB
model=GaussianNB()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

Calculate the score using cross validation

```python
from sklearn.model_selection import cross_val_score
cross_val_score(GaussianNB(),X_train,y_train,cv=5)
```
