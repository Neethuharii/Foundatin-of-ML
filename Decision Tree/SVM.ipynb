```python
import pandas as pd
from sklearn.datasets import load_iris
```

```python
iris=load_iris()
```

```python
iris.feature_names
```

```python
iris.target_names
```

```python
df = pd.DataFrame(iris.data,columns=iris.feature_names)
df.head()
```

```python
df['target'] = iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df[df.target==2].head()
```

```python
df['flower_name'] =df.target.apply(lambda x: iris.target_names[x])
df.head()
```

```python
df[45:55]
```

```python
df0=df[:50]
df1=df[50:100]
df2=df[100:]
```

```python
import matplotlib.pyplot as plt
```

Sepal length vs Sepal Width (Setosa vs Versicolor)

```python
plt.xlabel("Sepal length")
plt.ylabel("Sepal Width")
plt.scatter(df0["sepal length (cm)"],df0["sepal width (cm)"],marker='*',color='red')
plt.scatter(df1["sepal length (cm)"],df1["sepal width (cm)"],marker='+',color='green')
```

Petal length vs Pepal Width (Setosa vs Versicolor)

```python
plt.xlabel("Petal length")
plt.ylabel("Petal Width")
plt.scatter(df0["petal length (cm)"],df0["petal width (cm)"],marker='*',color='red')
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],marker='+',color='green')
```

```python
from sklearn.model_selection import train_test_split
```

```python
X=df.drop(['flower_name','target'],axis='columns')
y=df.target
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
```

```python
model=SVC()
```

```python
model.fit(X_train,y_train)
```

```python
model.score(X_test,y_test)
```

Tune parameters




1. Regularization (C)

```python
model_C=SVC(C=1)
model_C.fit(X_train,y_train)
model_C.score(X_test,y_test)
```

```python
model_g = SVC(gamma=10)
model_g.fit(X_train, y_train)
model_g.score(X_test, y_test)
```

```python
model_linear_kernal = SVC(kernel='linear')
model_linear_kernal.fit(X_train, y_train)
```

```python
model_linear_kernal.score(X_test, y_test)
```
